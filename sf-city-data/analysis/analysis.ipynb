# Overview of the San Francisco Public Safety Dataset

In this document we aim to provide an overview of the SF city public safety dataset, before we move on to our main aim -- to correlate Yelp restaurant ratings with public safety data and see whether:

1. areas with high crime are associated with low Yelp restaurant ratings, and
2. whether Yelp restaurant ratings can predict crime rate and severity

To begin, let's load the data.

```python
# some configuration stuff
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12
```

```python
import pandas as pd

DATA_PATH = '../sfpd-reported-incidents-2003-to-present'
NAME_F = 'sfpd_incident_%d.csv'
YEARS = range(2003, 2014 + 1)  # 2003 to 2014, inclusive

data = {}
for year in YEARS:
    name = NAME_F % year
    data[year] = pd.read_csv('%s/%s' % (DATA_PATH, name))
```

```python
print 'These years are available:\n%s' % data.keys()
print 'Each dataset has this many rows:\n%s' % [len(data20xx) for data20xx in data.values()]
```

```python
print "Here's what one looks like:"
data[2014].head()
```

All the fields are pretty self-explanatory. The differences between "Category" and "Descript" may be a bit vague, but we will look at this in just a bit. In general, "Category" is a broader idea than "Descript". What's important is the "X" and "Y" columns -- these correspond to longitude / latitude coordinates that map where the crime was reported to have happened. We will use this coordinate to query the Yelp API to find nearby restaurants to each crime.

Let's look at "Category" and "Descript".

### Category and Descript

```python
# just in 2014
print 'There are %d different types of "Category"s...' % len(data[2014]['Category'].unique())
print '... but a whopping %d different "Descript"s.' % len(data[2014]['Descript'].unique())
```

```python
data[2014]['Category'].unique()
```

The distribution of the categories looks like this:

```python
%matplotlib inline
import matplotlib.pyplot as plt

category_counts = pd.DataFrame({year: data[year]['Category'].value_counts() for year in YEARS})
category_counts = category_counts.fillna(value=0)
all_categories = category_counts.index
category_counts
```

```python
COLORS = ['r', 'yellow', 'g', 'b', 'purple', 'gray', 'cyan', 'magenta', 'orange', 'pink', 'white', 'black']

total = None
plots = []
for year in YEARS:
    heights = category_counts[year]
    plot = plt.bar(range(len(all_categories)), heights.tolist(),
                   color=COLORS[year-2003], align='center',
                   bottom=(total is not None and total.tolist()))
    plots.append(plot)
    if total is None:
        total = heights
    else:
        total += heights
plt.xticks(range(len(all_categories)), all_categories, rotation='vertical')
plt.legend([plot[0] for plot in plots], YEARS)
plt.show()
```

We noticed a few things here:

1. "LARCENY/THEFT" was by far the largest category of crime in SF.
2. There seems to have been more crime (reported) in the 2000s than in the 2010s. In addition, if we haphazardly assume that data collection was more difficult / less important to people in the 2000s, we can (haphazardly) conclude as well that crime has gone down significantly over the years.
3. In particular, "VEHICLE THEFT" has decreased signficantly over the years, from a peak of 18342 in 2005 to 6210 in 2013.

## Crime Density

```python
import numpy as np
import scipy.stats as stats

geo = data[2014][['X', 'Y']]
geo_array = geo.values.T

density = stats.gaussian_kde(geo_array)
xmin = geo_array[0].min()
xmax = geo_array[0].max()
ymin = geo_array[1].min()
ymax = geo_array[1].max()

mgrid = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
other = []
for i in xrange(100):
    other.append([])
    for j in xrange(100):
        coords = (mgrid[0][i][j], mgrid[1][i][j])
        other[i].append(density(coords)[0])
```
